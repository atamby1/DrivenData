# Blood Donations - A Deeper Look

In this exercise, I once again take a look at the blood donations dataset from DrivenData, but I go further than in my first exercise. I visualize the data and transform the variables and add new features to get an even better results.

Analysis inspired by http://www.philipkalinda.com/predicting_blood_donations.html

First, I set everything up and calculate average number of donations per month and add it to my feature set.

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, cross_validation, neighbors, metrics, linear_model, grid_search, ensemble
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
% matplotlib inline

train = pd.read_csv('9db113a1-cdbe-4b1c-98c2-11590f124dd8.csv', index_col = 0)
avg_train = (train['Months since First Donation'] - train['Months since Last Donation']) / train['Number of Donations']
train.insert(4, 'Average Number of Donations per Month', avg_train)
print(train.head(),'\n')

#Print null values
#train[train.isnull().any(axis=1)==True] # Check for Null values in training data

train.pivot_table(index = train.columns[-1], values = train.columns[:-1],aggfunc = np.mean)
print(np.corrcoef(train.transpose()))
```

Next, I see that Number of Donations is perfectly correlated with Total Volume Donated, so it does not provide any additional useful information. Each donation was 250cc, so TVD = ND * 250.

Below, I plot paired scatter plots for all of the variables to visualize the relationships and separate those who made donations in March 2007 and those who diddn't by color.

```python
sns.pairplot(train, hue = 'Made Donation in March 2007')
```

Next, I import the test data and add the Average Number of Donations per month variable.

```python
test = pd.read_csv('5c9fa979-5a84-45d6-93b9-543d1a0efc41.csv', index_col = 0)

avg_test = (test['Months since First Donation'] - test['Months since Last Donation']) / test['Number of Donations']
test.insert(4, 'Average Number of Donations per Month', avg_test)
print(test.head(),'\n')

#Print null values
#train[train.isnull().any(axis=1)==True] # Check for Null values in training data

print(np.corrcoef(test.transpose()))
```

Below, graph the densities of each variable to see how close the training set is to the test set.

```python
# Plot variable densities

f = plt.figure(figsize=(16,16))
ax1 = f.add_subplot(4,1,1)
ax2 = f.add_subplot(4,1,2)
ax3 = f.add_subplot(4,1,3)
ax4 = f.add_subplot(4,1,4)

ax1.set_title('Months since Last Donation')
sns.kdeplot(train['Months since Last Donation'], shade=True, cut=0, label='train_data',ax=ax1)
sns.kdeplot(test['Months since Last Donation'], shade=True, cut=0, label='test_data',ax=ax1)

ax2.set_title('Number of Donations')
sns.kdeplot(train['Number of Donations'], shade=True, cut=0, label='train_data',ax=ax2)
sns.kdeplot(test['Number of Donations'], shade=True, cut=0, label='test_data',ax=ax2)

ax3.set_title('Average Number of Donations per Month')
sns.kdeplot(train['Average Number of Donations per Month'], shade=True, cut=0, label='train_data',ax=ax3)
sns.kdeplot(test['Average Number of Donations per Month'], shade=True, cut=0, label='test_data',ax=ax3)

ax4.set_title('Months since First Donation')
sns.kdeplot(train['Months since First Donation'], shade=True, cut=0, label='train_data',ax=ax4)
sns.kdeplot(test['Months since First Donation'], shade=True, cut=0, label='test_data',ax=ax4)

plt.show()
```

Both data sets seem to be distributed pretty closely, but I want to see if taking the log gives me an even closer distribution.

```python
train['log Months since Last Donation'] = train['Months since Last Donation'].apply(lambda x: np.log(x+1))
test['log Months since Last Donation'] = test['Months since Last Donation'].apply(lambda x: np.log(x+1))

train['log Number of Donations'] = train['Number of Donations'].apply(lambda x: np.log(x))
test['log Number of Donations'] = test['Number of Donations'].apply(lambda x: np.log(x))

train['log Average Number of Donations per Month'] = train['Average Number of Donations per Month'].apply(lambda x: np.log(x))
test['log Average Number of Donations per Month'] = test['Average Number of Donations per Month'].apply(lambda x: np.log(x))

train['log Months since First Donation'] = train['Months since First Donation'].apply(lambda x: np.log(x))
test['log Months since First Donation'] = test['Months since First Donation'].apply(lambda x: np.log(x))

# Plot variable densities

f = plt.figure(figsize=(16,16))
ax1 = f.add_subplot(4,1,1)
ax2 = f.add_subplot(4,1,2)
ax3 = f.add_subplot(4,1,3)
ax4 = f.add_subplot(4,1,4)

ax1.set_title('log Months since Last Donation')
sns.kdeplot(train['log Months since Last Donation'], shade=True, cut=0, label='train_data',ax=ax1)
sns.kdeplot(test['log Months since Last Donation'], shade=True, cut=0, label='test_data',ax=ax1)

ax2.set_title('log Number of Donations')
sns.kdeplot(train['log Number of Donations'], shade=True, cut=0, label='train_data',ax=ax2)
sns.kdeplot(test['log Number of Donations'], shade=True, cut=0, label='test_data',ax=ax2)

ax3.set_title('log Average Number of Donations per Month')
sns.kdeplot(train['log Average Number of Donations per Month'], shade=True, cut=0, label='train_data',ax=ax3)
sns.kdeplot(test['log Average Number of Donations per Month'], shade=True, cut=0, label='test_data',ax=ax3)

ax4.set_title('log Months since First Donation')
sns.kdeplot(train['log Months since First Donation'], shade=True, cut=0, label='train_data',ax=ax4)
sns.kdeplot(test['log Months since First Donation'], shade=True, cut=0, label='test_data',ax=ax4)

plt.show()
```

The log of the variables seem to have a closer distribution, so maybe I will use those as my feature variables.

```python
X = np.array(train[['log Months since Last Donation','log Number of Donations', 
                            'Average Number of Donations per Month','log Months since First Donation']])
print(X[:5])
y = np.array(train['Made Donation in March 2007'])
print(y[:5])
```

```python
n = len(X)

logit = linear_model.LogisticRegression(n_jobs = -1)
knn = neighbors.KNeighborsClassifier(n_neighbors = 30, n_jobs=-1)
rf = ensemble.RandomForestClassifier(n_jobs = -1)
gbt = ensemble.GradientBoostingClassifier()
bag = ensemble.BaggingClassifier(n_jobs = -1)
xgb = xgb.XGBClassifier()
```

```python
lr_param_grid = {'C': [3,2,1,0.5,0.1,0.05,0.01,0.001]}
lr_grid = grid_search.GridSearchCV(logit, lr_param_grid, cv=10)
lr_grid.fit(X, y)
lr_grid.best_params_
```

```python
logit = linear_model.LogisticRegression(C = 1, n_jobs = -1)
```

```python
def classifier(alg, features, y, k=1):
    preprocessing.scale(features)
    knn = neighbors.KNeighborsClassifier(n_neighbors = k, n_jobs = -1)

    loocv = cross_validation.LeaveOneOut(n) # Create n training and test sets
    kfcv = cross_validation.KFold(n, n_folds = 10)
    error_rates = [] # Initialize vector of error rate for each iteration of CV
    probs = np.empty([0,2]) # Initialize nx2 matrix of prediction probabilities
    for train, test in kfcv:
        alg.fit(X[train,:], y[train]) # fit algorithm on training set
        error = 1 - alg.score(X[test,:], y[test]) # Find the error on the test set
        error_rates.append(error) # Create n-vector of error rates
        alg_train_pp = alg.predict_proba(X[test,:]) # Calculate prediction probabilities
        probs = np.append(probs, alg_train_pp, axis = 0) # Create nx2 matrix of prediction probabilities
    #gen_error_rate = sum(error_rates) / n # Gen. error is mean of n error rates in CV
    #print('Generalization Error =', gen_error_rate*100, '%')
    log_loss = metrics.log_loss(y, probs) # Calculate log-loss
    print('Accuracy = ', 1 - error)
    print('Log Loss Score =', log_loss)
    fpr, tpr, thresholds = metrics.roc_curve(y, probs[:,1])
    print('AUC = ', metrics.auc(fpr, tpr))
    print()
    return fpr, tpr
```

```python
lfpr, ltpr = classifier(logit, X, y)
kfpr, ktpr = classifier(knn, X, y, 40)
rfpr, rtpr = classifier(rf, X, y)
gfpr, gtpr = classifier(gbt, X, y)
bfpr, btpr = classifier(bag, X, y)
xfpr, xtpr = classifier(xgb, X, y)
```

```python
plt.figure(figsize=(16,6)); plt.xlim([-0.2, 1.3]); plt.ylim([0, 1.1])
plt.title('ROC Curves'); plt.xlabel('False Positive Rate (FPR)'); plt.ylabel('True Positive Rate (TPR)')
plt.plot(lfpr, ltpr, label='Logistic Regression')
plt.plot(kfpr, ktpr, label='k-Nearest Neighbors')
plt.plot(rfpr, rtpr, label='Random Forest')
plt.plot(gfpr, gtpr, label='Gradient Boosting Classifier')
plt.plot(bfpr, btpr, label='Bagging Classifier')
plt.plot(xfpr, xtpr, label='XGBoost Classifier')
plt.legend()
plt.show()
```

```python
X2 = np.array(test[['log Months since Last Donation','log Number of Donations', 
                            'Average Number of Donations per Month','log Months since First Donation']])

probability = logit.predict_proba(X2)
p = []
for i in range(len(X2)):
    p.append(probability[i,1])

p = pd.DataFrame(p)
p['index'] = test.index
p = p[['index', 0]]
p.columns = ['', 'Made Donation in March 2007']
#print(p)
p.to_csv('prediction2.csv', index=False)
```
